```python
string_columns = h1.select_dtypes(include=['object']).columns
```

```python
top_frequent_values = {}
for col in string_columns:
    value_counts = h1[col].value_counts()
    top_values = value_counts.head(4)  # This gives the top 4 most frequent values
    top_frequent_values[col] = top_values
```

```python
for col, top_values in top_frequent_values.items():
    print(f"Column: {col}\n{top_values}\n")
```
