### 2. Exploring Relationships

#### Top AssignedRoomType

```python
# Visualizing the top 15 most reordered products

# Counting the number of bookings for each room type
room_type_counts = h1['AssignedRoomType'].value_counts()

# Creating the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=room_type_counts.index, y=room_type_counts.values)
plt.title('Bookings by Assigned Room Type')
plt.xlabel('Room Type')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.show()
```

#### Headmap

```python
# Create a subset excluding ID and boolean (1 and 0) columns

sub = h1[['ADR', 'ArrivalDateWeekNumber', 'ArrivalDateDayOfMonth', 'LeadTime', 'StaysInWeekendNights', 'StaysInWeekNights', 'Adults', 'Children','Babies', 'RequiredCarParkingSpaces', 'TotalOfSpecialRequests', 'PreviousCancellations', 'PreviousBookingsNotCanceled', 'BookingChanges']]
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,9))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot=True, cmap="Reds", ax = ax)
```

## Discuss what the coefficients in the plot mean in terms of the relationships between the variables.

Overall, we don't have strong coefficients. Bookings only have a high correlations between weekend-days and weekdays. This means that the booking duration include more often weekend-days AND weekdays. But this is due to the fact that a booking duration over 5 days always has to include at least 1 weekend-day, creating naturally a correlation.

#### Categorical Plots

Creating a categorical variable that splits the Average Daily Rate "ADR" column into categories

```python
# Use a histogram to visualize the distribution of the variables. 
# This way, you can determine sensible categories for the average daily rate. 
# You don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution

sns.histplot(h1['ADR'], bins = 50, kde = False, binrange = (0, 400))

# Customize the x-axis ticks
plt.xticks([0, 25, 50, 75, 100, 150, 200, 250, 300, 350, 400])

# Show the plot
plt.show()
```

## Interpretion of the results.

The resort hotel most bookings costs around 40€ - 100€ per night.

```python
h1.loc[h1['ADR'] < 40, 'Price category'] = 'Low ADR'
```

```python
h1.loc[(h1['ADR'] >= 40) & (h1['ADR'] < 100), 'Price category'] = 'Middle ADR'
```

```python
h1.loc[h1['ADR'] >= 100, 'Price category'] = 'High ADR'
```

```python
h1['Price category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(x="RequiredCarParkingSpaces", y="ADR", hue="Price category", data=h1)
```

The price category helps to see that no booking with a low ADR has more than 2 parking spots.
